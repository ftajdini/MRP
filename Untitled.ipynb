```python
# Scatter plot for missing values

def scatter_plot_with_missing(data, x, y, xlabel, ylabel):
    plt.figure(figsize=(10, 6))

    non_missing_data = data.dropna(subset=[x, y])
    sns.scatterplot(x=non_missing_data[x], y=non_missing_data[y], color='blue', label='Non-missing')

    missing_data_x = data[data[x].isna()]
    sns.scatterplot(x=[0]*len(missing_data_x), y=missing_data_x[y], color='red', label='Missing')

    missing_data_y = data[data[y].isna()]
    sns.scatterplot(x=missing_data_y[x], y=[0]*len(missing_data_y), color='red')
    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(f'{xlabel} vs {ylabel}')
    plt.legend()
    plt.show()

scatter_plot_with_missing(df, 'ALP', 'CHOL', 'ALP', 'CHOL')
scatter_plot_with_missing(df, 'ALB', 'PROT', 'AB', 'PROT')
```

```python
X = df.drop(columns=['Category'])
y = df['Category']

# Define the function for k-fold cross-validation
def evaluate_model_kfold(model, X, y, n_splits=5, is_classification=True):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    accuracy_list = []
    precision_list = []
    recall_list = []
    f1_list = []
    confusion_matrices = []

    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        if isinstance(model, Sequential):
            # Prepare data for RNN
            X_train_rnn = np.expand_dims(X_train.values, axis=2)
            X_test_rnn = np.expand_dims(X_test.values, axis=2)
            y_train_rnn = to_categorical(y_train)
            y_test_rnn = to_categorical(y_test)
            model.fit(X_train_rnn, y_train_rnn, epochs=50, batch_size=32, verbose=0)
            y_pred_prob = model.predict(X_test_rnn)
            y_pred = np.argmax(y_pred_prob, axis=1)
        else:
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
        
        if is_classification:
            accuracy = accuracy_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred)
            recall = recall_score(y_test, y_pred)
            f1 = f1_score(y_test, y_pred)
            cm = confusion_matrix(y_test, y_pred)
            
            accuracy_list.append(accuracy)
            precision_list.append(precision)
            recall_list.append(recall)
            f1_list.append(f1)
            confusion_matrices.append(cm)

    if is_classification:
        print(f'Model: {model.__class__.__name__}')
        print(f'Average Accuracy: {np.mean(accuracy_list)}')
        print(f'Average Precision: {np.mean(precision_list)}')
        print(f'Average Recall: {np.mean(recall_list)}')
        print(f'Average F1 Score: {np.mean(f1_list)}')
        print('\n')
        
       
        plt.figure(figsize=(8, 6))
        sns.heatmap(confusion_matrices[-1], annot=True, fmt='d', cmap='Blues')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title(f'Confusion Matrix for {model.__class__.__name__}')
        plt.show()
    else:
        print(f'Model: {model.__class__.__name__}')
        print(f'Average MSE: {np.mean(mse_list)}')
        print('\n')
```

```python
# Select features and target
important_features = ['ALT_AST_Ratio', 'Age_AST_Interaction', 'AST', 'GGT_ALT_Ratio', 'ALT', 'GGT', 'Age_ALT_Interaction']
X = df[important_features]
y = df['Category']


# Apply PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_scaled)

# Function to evaluate models
def evaluate_model_kfold(model, X, y, n_splits=5):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    accuracy_list = []
    precision_list = []
    recall_list = []
    f1_list = []
    confusion_matrices = []
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)

    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        if isinstance(model, Sequential):
            # Prepare data for RNN
            X_train_rnn = np.expand_dims(X_train, axis=2)
            X_test_rnn = np.expand_dims(X_test, axis=2)
            y_train_rnn = to_categorical(y_train)
            y_test_rnn = to_categorical(y_test)
            model.fit(X_train_rnn, y_train_rnn, epochs=50, batch_size=32, verbose=0)
            y_pred_prob = model.predict(X_test_rnn)
            y_pred = np.argmax(y_pred_prob, axis=1)
            y_pred_prob = y_pred_prob[:, 1]  # Only keep the probability for the positive class
        else:
            model.fit(X_train, y_train)
            y_pred_prob = model.predict_proba(X_test)[:, 1]
            y_pred = model.predict(X_test)
        
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        cm = confusion_matrix(y_test, y_pred)
            
        accuracy_list.append(accuracy)
        precision_list.append(precision)
        recall_list.append(recall)
        f1_list.append(f1)
        confusion_matrices.append(cm)

        # Compute ROC curve and area the curve
        fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
        tprs.append(np.interp(mean_fpr, fpr, tpr))
        roc_auc = auc(fpr, tpr)
        aucs.append(roc_auc)
    
    print(f'Model: {model.__class__.__name__}')
    print(f'Average Accuracy: {np.mean(accuracy_list)}')
    print(f'Average Precision: {np.mean(precision_list)}')
    print(f'Average Recall: {np.mean(recall_list)}')
    print(f'Average F1 Score: {np.mean(f1_list)}')
    print('\n')
        
    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(confusion_matrices[-1], annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix for {model.__class__.__name__}')
    plt.show()
        
    # Plot ROC curve
    plt.figure(figsize=(10, 8))
    plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='gray', alpha=0.8)
        
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    plt.plot(mean_fpr, mean_tpr, color='b',
             label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
             lw=2, alpha=0.8)
        
    std_tpr = np.std(tprs, axis=0)
    tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
    tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
    plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='gray', alpha=0.2,
                     label=r'$\pm$ 1 std. dev.')
        
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic (ROC) Curve for {model.__class__.__name__}')
    plt.legend(loc="lower right")
    plt.show()
```

```python
# Evaluate SVM model
svm_model = SVC(kernel='linear', probability=True, random_state=42)
evaluate_model_kfold(svm_model, X_pca, y)
```

```python
# Select features for PCA
X = df.drop(columns=['Category'])

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_pca)

# Add cluster labels to the dataframe
df['Cluster'] = clusters

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='viridis')
plt.title('K-means Clustering with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
# Analyze cluster characteristics
cluster_features = df.groupby('Cluster').mean()
print(cluster_features)
```

```python
# Calculate the silhouette score
sil_score = silhouette_score(X_pca, clusters)

print(f'Silhouette Score: {sil_score}')


from sklearn.metrics import silhouette_samples
import numpy as np


silhouette_vals = silhouette_samples(X_pca, clusters)


plt.figure(figsize=(10, 6))
y_lower, y_upper = 0, 0
yticks = []
for i in range(3):  
    cluster_silhouette_vals = silhouette_vals[clusters == i]
    cluster_silhouette_vals.sort()
    y_upper += len(cluster_silhouette_vals)
    plt.barh(range(y_lower, y_upper), cluster_silhouette_vals, height=1.0)
    yticks.append((y_lower + y_upper) / 2)
    y_lower += len(cluster_silhouette_vals)

plt.axvline(x=sil_score, color="red", linestyle="--")
plt.yticks(yticks, [f'Cluster {i}' for i in range(3)])
plt.xlabel('Silhouette Coefficient')
plt.ylabel('Cluster')
plt.title('Silhouette Analysis for K-means Clustering')
plt.show()
```
